```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

alpha = 1
beta_param = 1
p = np.linspace(0, 1, 100)

plt.plot(p, beta.pdf(p, alpha, beta_param), 'r-', lw=2, label='Prior (Beta(1,1))')
plt.xlabel('p')
plt.ylabel('Density')
plt.title('Uniform Prior for p')
plt.show()
```

```python
alpha = 6
beta_param = 4
p = np.linspace(0, 1, 100)

plt.plot(p, beta.pdf(p, alpha, beta_param), 'r-', lw=2, label='Prior (Beta(6,4))')
plt.xlabel('p')
plt.ylabel('Density')
plt.title('Prior Centered Around 0.6')
plt.show()
```

```python
alpha_post = alpha + 120
beta_post = beta_param + 80
p = np.linspace(0, 1, 100)

plt.plot(p, beta.pdf(p, alpha_post, beta_post), 'b-', lw=2, label='Posterior (Beta(126,84))')
plt.xlabel('p')
plt.ylabel('Density')
plt.title('Posterior Distribution for p')
plt.show()
```

```python
alpha_prior = 30
beta_prior = 70
alpha_post = alpha_prior + 120
beta_post = beta_prior + 80

plt.figure(figsize=(10, 5))

# Prior
plt.subplot(1, 2, 1)
plt.plot(p, beta.pdf(p, alpha_prior, beta_prior), 'r-', lw=2, label='Prior (Beta(30,70))')
plt.xlabel('p')
plt.ylabel('Density')
plt.title('Prior Reflecting Belief Around 0.3')

# Posterior
plt.subplot(1, 2, 2)
plt.plot(p, beta.pdf(p, alpha_post, beta_post), 'b-', lw=2, label='Posterior (Beta(150,150))')
plt.xlabel('p')
plt.ylabel('Density')
plt.title('Posterior Distribution')

plt.tight_layout()
plt.show()
```
