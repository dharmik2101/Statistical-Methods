```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def clean_data(data):
    data = data.dropna(subset=['loss_percentage', 'loss_quantity'])
    data['loss_percentage'] = pd.to_numeric(data['loss_percentage'], errors='coerce')
    data['loss_quantity'] = pd.to_numeric(data['loss_quantity'], errors='coerce')
    data = data.drop_duplicates()
    return data

def plot_normal_distribution(data, column):
    """
    Plot the fitted normal distribution for a given column.
    """
    mean_val = np.mean(data[column])
    std_dev_val = np.std(data[column])
    x = np.linspace(mean_val - 4 * std_dev_val, mean_val + 4 * std_dev_val, 100)
    y = norm.pdf(x, mean_val, std_dev_val)
    
    plt.plot(x, y, label="Fitted Normal Distribution")
    plt.title(f"Normal Distribution: {column}")
    plt.xlabel(column)
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    data = pd.read_csv(r"C:\Users\dshah\OneDrive\Documents/cleaned_FAO_food_loss.csv")
    cleaned_data = clean_data(data)
    plot_normal_distribution(cleaned_data, "loss_percentage")
```
